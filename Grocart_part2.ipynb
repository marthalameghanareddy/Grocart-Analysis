```
import numpy as np
import pandas as pd
```

```
orders_data=pd.read_csv('/content/drive/My Drive/Grocart_Project/orders.csv')
```

```
df_dow=pd.read_csv('/content/drive/My Drive/Grocart_Project/df_dow.csv')
df_dow.head()
```

```
df_hour=pd.read_csv('/content/drive/My Drive/Grocart_Project/df_hour.csv')
df_hour.head()
```

```
df_products_merged=pd.read_csv('/content/drive/My Drive/Grocart_Project/df_products_merged.csv')
df_products_merged.head()
```

```
df_user=pd.read_csv('/content/drive/My Drive/Grocart_Project/df_user.csv')
df_user.head()
```

```
df_user_product=pd.read_csv('/content/drive/My Drive/Grocart_Project/df_user_product.csv')
df_user_product.head()
```

```
merge_1 = pd.merge(df_user,df_user_product,on = 'user_id',how = 'inner')
merge_1.head()
```

```
final_merge = pd.merge(merge_1, df_products_merged, on = 'product_id', how = 'inner')
final_merge.sample(10)
```

```
orders_last = orders_data.loc[((orders_data.eval_set == 'train') | (orders_data.eval_set == 'test')), ['user_id', 'eval_set', 'order_id', 'order_dow', 'order_hour_of_day', 'days_since_prior_order']]
orders_last.head()
```

```
final = final_merge.merge(orders_last, on='user_id', how='left')
final.to_csv('/content/drive/My Drive/Grocart_Project/final.csv', index = False)
final.head()
```
